```python
# import dependency
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to list of strings for each month
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# matlab
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Using ax.plot
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create plot adding labels, title and legend
plt.plot(x_axis, y_axis, label = 'Boston')
# Create labels
plt.xlabel('Date')
plt.ylabel('Fare($)')
#set y limit
plt.ylim(0, 45)
# Create a title.
plt.title('PyBer Fare by Month')
# add legend
plt.legend()
```

```python
# create plot
plt.plot(x_axis, y_axis, marker='*', color='blue',linewidth=2, label='Boston')
# Create labels
plt.xlabel('Date')
plt.ylabel('Fare($)')
#set y limit
plt.ylim(0, 45)
# Create a title.
plt.title('PyBer Fare by Month')
# create grid
plt.grid()
# add legend
plt.legend()
plt.xlim('Jan','Jun')
```

```python
# create plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='green', marker='D',label='Boston')
# Create labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#set y limit
ax.set_ylim(0, 45)
# Create a title.
ax.set_title('PyBer Fare by Month')
# create grid
ax.grid()
# add legend
plt.legend()
```

## Create Scatter Plots and Bubble Charts

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# skill drill
plt.scatter(y_axis,x_axis,c='red',label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
import pandas as pd
plt.scatter(x_axis, y_axis, s=pd.Series(y_axis)*3)
```

```python
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis,s = [i* 3 for i in y_axis])
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
```

```python
# skill drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,c='skyblue',edgecolors = 'k',alpha=0.8, linewidth=2,s = [i* 5 for i in y_axis],label='Chicago')
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.legend()
ax.invert_yaxis()
```

```python
# Pie chart
plt.pie(y_axis, labels=x_axis)
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis,explode=explode_values,autopct='%1.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,labels=x_axis,
        colors=colors,
        autopct='%1.1f%%',
        explode=explode_values
       )
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
       labels=x_axis,
)
```

```python
#skill drill
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
       labels=x_axis,
       explode = [0.2 if i >sorted(y_axis)[-3] else 0 for i in y_axis],
       autopct='%1.1f%%',
       shadow=True,
       startangle=270-y_axis[0]/sum(y_axis)*360/2,
       colors=colors
)
ax.invert_yaxis()
```
